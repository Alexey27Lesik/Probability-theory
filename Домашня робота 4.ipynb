## 3 Різні параметри

### 3.1 Біноміальний розподіл

Зобразіть графічно функцію щільності, функцію розподілу та гістограму $10000$ випадкових величин біноміального розподілу з параметрами:
1. $n = 15, p = 0.8$
2. $n = 15, p = 0.2$
3. $n = 20, p = 0.4$
4. $n = 20, p = 0.6$
5. $n = 7, p = 0.5$
6. $n = 25, p = 0.5$
7. $n = 100, p = 0.5$

```python
from scipy.stats import binom
p = [[15, 0.8], [15, 0.2], [20, 0.4], [20, 0.6], [7, 0.5], [25, 0.5], [100, 0.5]]
for n, i in p:
    x = np.arange(0, n+1) 
    y1 = binom.pmf(x, n, i)
    plt.subplot(1, 3, 1)
    plt.stem(x, y1)
    
    y2 = binom.cdf(x, n, i)
    plt.subplot(1, 3, 2)
    plt.plot(x, y2, drawstyle='steps-post')
    
    plt.title('n = {}, p = {}'.format(n, i))
    
    y3 = binom.rvs(n, i, size=10000)
    plt.subplot(1, 3, 3)
    plt.hist(y3)
    plt.show()
```

## 4 Апроксимація біноміального розподілу розподілом Пуассона

Покажіть схожість біноміального розподілу та розподілу Пуассона (їх функцій щільності) з відповідними параметрами.
Наприклад:
- $n = 100, p = 0.05$
- $n = 100, p = 0.1$
- $n = 100, p = 0.15$
- $n = 100, p = 0.2$
- $n = 100, p = 0.35$
- $n = 100, p = 0.5$
- $n = 100, p = 0.75$

```python
from scipy.stats import poisson
p = [[100,0.05],[100, 0.1], [100,0.15], [100,0.2], [100,0.35], [100,0.5], [100,0.75]]

for n, i in p:
    x = np.arange(0, n+1)
    y1 = binom.pmf(x, n, i)
    y2 = poisson.pmf(x, i*n)
    plt.stem(x, y1,'r' ,label='binom density') 
    plt.stem(x, y2,'b', label='poisson density')
    plt.legend()
    plt.title('n = {}, p = {}'.format(n, i))
    plt.show()
```
