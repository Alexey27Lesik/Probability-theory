## 1 Задача 1.3.12
Знайти ймовiрнiсть того, що серед трьох цифр, кожна з яких вибрана навмання, буде лише 1, 2, 3 рiзних.

Обчисліть теоретичну (повним перебором) та еміричну (симулюванням $100000$ експериментів) імовірності.

```python
# Місце для Вашого коду
from itertools import product
import numpy as np
 
def teor_prob(p):
    a = list(product(range(10), repeat = 3))
    b = []
    for x in a:
        if len(set(x)) == p:
            b.append(x)
    return len(b)/10**3

def emp_prob(p):
    a = []
    b = []
    for i in range(10**3):
        l = list(np.random.choice(range(10), size=3))
        a.append(l)
    for x in a:
        if len(set(x)) == p:
            b.append(x)
    return len(b)/10**3


for i in [1, 2, 3]:
    print("Theor prob when {} different numbers {}".format(i, teor_prob(i)))
    print("Empir prob when {} different numbers {}".format(i, emp_prob(i)))
```

# 2 Задача 1.3.14
З послiдовностi чисел $1, 2, . . . , n$ вибирають навмання $k$ рiзних чисел.
Яка ймовiрнiсть того, що:
1. кожне з вибраних чисел кратне даному числу $p$;
2. кожне з вибраних чисел кратне хоча б одному з двох взаємно простих чисел $p$ i $q$;
3. серед вибраних чисел є хоча б одне кратне $p$?

Напишіть  відповідні функції для обрахунку теоретичної (повним перебором) та еміричної (симулюванням $100000$ експериментів) імовірностей в залежності від параметрів $n, k, p, q$.
Виведіть результат для
- $n = 25, k = 5, p = 3, q = 4$;
- $n = 25, k = 10, p = 3, q = 4$.

```python
import numpy as np
import itertools

def task_a(n, k, p):
    a = list(itertools.combinations(range(1, n + 1), k))
    b = []
    bool = False
    for i in range(len(a)):
        for j in range(k):
            if a[i][j] % p == 0:
                bool = True
            else:
                bool = False
                break
        if bool:
            b.append(a[i][j])
    return len(b)/(len(a))

def emp_a(n, k, p):
    a = list(itertools.combinations(range(1, n + 1), k))
    b = []
    for _ in range(100000):
        a = list(np.random.choice(list(range(1, n + 1)), size=k, replace=False))
        bool = False
        for i in a:
            if i % p == 0:
                bool = True
            else:
                bool = False
                break
        if bool:
            b.append(i)

    return len(b)/ 100000

def task_b(n, k, p, q):
    a = list(itertools.combinations(range(1, n + 1), k))
    b = []
    bool = False
    for i in range(len(a)):
        for j in range(k):
            if a[i][j] % p == 0 or a[i][j] % q == 0:
                bool = True
            else:
                bool = False
                break
        if bool:
            b.append(a[i][j])
    return len(b)/(len(a))

def emp_b(n, k, p, q):
    b = []
    for _ in range(100000):
        a = list(np.random.choice(list(range(1, n + 1)), size=k, replace=False))
        bool = False
        for i in a:
            if i % p == 0 or i % q == 0:
                bool = True
            else:
                bool = False
                break
        if bool:
            b.append(i)

    return len(b)/ 10**5

def task_c(n, k, p):
    a = list(itertools.combinations(range(1, n + 1), k))
    b = []
    for i in range(len(a)):
        for j in range(k):
            if a[i][j] % p == 0:
                b.append(a[i][j])
                break
    return len(b)/(len(a))

def emp_c(n, k, p):
    b = []
    for _ in range(100000):
        a = list(np.random.choice(list(range(1, n + 1)), size=k, replace=False))
        for i in a:
            if i % p == 0:
                b.append(i)
                break
    return len(b) / 10**5


val = [[25, 5, 3, 4], [25, 10, 3, 4]]
print("Task_2a")

for i in val:
    print("Theor prob n={}, k={}, p={} is {}".format(i[0], i[1], i[2], '%.30f' % task_a(i[0], i[1], i[2])))
    print("Emp prob is {}".format('%.30f' % emp_a(i[0], i[1], i[2])))
    
print("Task_2b")
for i in val:
    print("Theor prob n={}, k={}, p={}, q={} is {}".format(i[0], i[1], i[2], i[3], '%.30f' % task_b(i[0], i[1], i[2], i[3])))
    print("Emp prob {}".format('%.30f' % emp_b(i[0], i[1], i[2], i[3])))


print("Task_2c")

for i in val:
    print("Theor prob n={}, k={}, p={} is {}".format(i[0], i[1], i[2], '%.30f' % task_c(i[0], i[1], i[2])))
    print("Emp prob is {}".format('%.30f' % emp_c(i[0], i[1], i[2])))
```

# 3 Задача 1.4.6
В коморi знаходяться $n$ пар черевикiв.
З них випадковим чином вибираються $2k$ черевикiв.
Яка ймовiрнiсть того, що серед вибраних черевикiв:
1. вiдсутнi парнi;
2. є рiвно одна комплектна пара;
3. є двi комплектнi пари?

Напишіть  відповідні функції для обрахунку теоретичної (повним перебором) та еміричної (симулюванням $100000$ експериментів) імовірностей в залежності від параметрів $n, k$.
Виведіть результат для
- $n = 8, k = 4$;
- $n = 13, k = 5$.

```python
import itertools
import numpy as np
def condition(comb):
    res = 0
    res1 = 0
    res2 = 0
    lenth = len(comb)
    for i in comb:
        if len(i) == len(set(i)):
            res += 1
        elif len(i)  == len(set(i)) + 1:
            res1 += 1
        elif len(i) == len(set(i)) + 2:
            res2 += 1
    return res / lenth, res1 / lenth, res2 / lenth

val = [[8,4],[13,5]]
for i in val:
    shoes = [x for x in range(1, i[0] + 1)] * 2

    comb = list(itertools.combinations(shoes, r=2 * i[1]))
    z, o, t = condition(comb)
    
    n = 100000
    lst = []
    for j in range(n):
        a = np.random.choice(shoes, size=2 * i[1], replace=False)
        lst.append(a)
    z_e, o_e, t_e = condition(lst)
    print(i)
    print("Theor prob _ 1) ", z)
    print("Empir prob _ 1) ", z_e)
    print("Theor prob _ 2) ", o)
    print("Empir prob _ 2) ", o_e)
    print("Theor prob _ 3) ", t)
    print("Empir prob _ 3) ", t_e)
```

# 4 Задача 1.2.15
Нехай $\Omega = {1, 2, \ldots, 2n}$.
Всiм числам приписанi ймовiрностi, пропорцiйнi логарифмам цих чисел.
Знайти цi ймовiрностi.
Знайти ймовiрнiсть того, що в результатi експерименту з’явиться:
1. парне число;
2. непарне число.

Напишіть функцію для обрахунку еміричної (симулюванням $100000$ експериментів) імовірності в залежності від параметра $n$.
Виведіть результат для
- $n = 10$;
- $n = 25$.

```python
import numpy as np

def funk(n):
    num = list(range(1, 2 * n + 1))
    logs = []
    for i in num:
        a = np.log(i)
        logs.append(a)
    c = (1/sum(logs))
    lst = []
    for i in logs:
        b = i*c
        lst.append(b)
    sum_even = 0
    sum_odd = 0
    for i in lst[1::2]:
        sum_even += i
    for j in lst[1::2]:
        sum_odd += j
    return "Even ", sum_even, "Odd ", sum_odd

def Emp(n):
    even = 0
    odd = 0
    for i in range(100000):
        num = np.random.randint(0, 2*n + 1)
        if num % 2 == 0:
            even += 1
        else:
            odd += 1
    return "Even ", even / 10**5, "Odd ", odd / 10**5

for n in [10, 25]:
    print( "n =", n,"Theor prob ", funk(n))
    print("Emp prob ", Emp(n))
```
